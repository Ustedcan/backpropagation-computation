Definir las variables como instancias de Value de micrograd, asegurando que cada
variable tenga una etiqueta representativa.

La ecuación a implementar es:  

$$ L = \left( (a \times b + c)^2 + d + e \times f \right) \times g $$


Donde: 

$$
a = 2, \quad b = 3, \quad c = 1, \quad d = 1, \quad e = 4, \quad f = 2, \quad g = 5

```python
# a=2, b=3, c=1, d=1, e=4, f=2, g=5
# 𝐿 = ((𝑎 × 𝑏 + 𝑐)^2 + 𝑑 + 𝑒 × 𝑓) × g)

a = Value(2.0, label='a')
print(a)
b = Value(3.0, label='b')
print(b)
c = Value(1.0, label='c')
print(c)
d = Value(1.0, label='d')
print(d)
e = Value(4.0, label='e')
print(e)
f = Value(2.0, label='f')
print(f)
g = Value(5.0, label='g')
print(g)

# Definimos los calculos
h = a * b; h.label = 'h'
print(h)

i = h + c; i.label = 'i'
print(i)

j = i ** 2; j.label = 'j'
print(j)

k = e * f; k.label = 'k'
print(k)

l = k + d; l.label = 'l'
print(l)

m = j + l; m.label = 'm'
print(m)

L = m * g; L.label = 'L'
print(L)
```

Construir la expresión matemática paso a paso en código, asegurando que cada
operación intermedia tenga una etiqueta clara.

```python
# Definimos los valores
a = Value(2.0, label='a')
b = Value(3.0, label='b')
c = Value(1.0, label='c')
d = Value(1.0, label='d')
e = Value(4.0, label='e')
f = Value(2.0, label='f')
g = Value(5.0, label='g')

# Se realizan las operaciones matemáticas que componen L, son los mismos calculos
# que se realizaron en el paso anterior.
h = a * b; h.label = 'h'
i = h + c; i.label = 'i'
j = i ** 2; j.label = 'j'
k = e * f; k.label = 'k'
l = k + d; l.label = 'l'
m = j + l; m.label = 'm'
L = m * g; L.label = 'L'
```

Realizar la backpropagation para obtener los gradientes de todas las variables
involucradas en la ecuación

```python
# Se calcula las derivadas para todos los casos a traves de la función backward()
L.backward()
```

```python
# Imprimimos
# Para todos los datos
print(f"Gradiente dL/db: {b.grad}")
print(f"Gradiente dL/dc: {c.grad}")
print(f"Gradiente dL/dd: {d.grad}")
print(f"Gradiente dL/de: {e.grad}")
print(f"Gradiente dL/df: {f.grad}")
print(f"Gradiente dL/dg: {g.grad}")

# Para la operaciones
print(f"Gradiente dL/dh: {h.grad}")
print(f"Gradiente dL/di: {i.grad}")
print(f"Gradiente dL/dj: {j.grad}")
print(f"Gradiente dL/dk: {k.grad}")
print(f"Gradiente dL/dl: {l.grad}")
print(f"Gradiente dL/dm: {m.grad}")

# El resultado final de L respecto a a (dL/da)
print(f"Valor final de L: {L}")
print(f"Gradiente dL/da: {a.grad}")
```
